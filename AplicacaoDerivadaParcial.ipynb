#Exemplo de uma função que calcula o lucro de um FIDC, usando fórmulas financeiras básicas

Parâmetros:
  - valor_investido: Valor investido no FIDC.
  - taxa_retorno: Taxa de retorno dos direitos creditórios.
  - taxa_administracao: Taxa de administração do FIDC (em percentual).

Retorna:
  - lucro: Lucro do FIDC após descontar a taxa de administração.

```python
def calcular_lucro_fidc(valor_investido, taxa_retorno, taxa_administracao):
    retorno_bruto = valor_investido * taxa_retorno
    desconto_administracao = valor_investido * (taxa_administracao / 100)
    lucro = retorno_bruto - desconto_administracao
    return lucro
```

```python
# Exemplo de uso da função.

valor_investido = 100000
taxa_retorno = 0.08
taxa_administracao = 1.5

lucro = calcular_lucro_fidc(valor_investido, taxa_retorno, taxa_administracao)
print(f"Lucro estimado do FIDC: R$ {lucro:.2f}")
```

Lembrando que essa é uma função simplificada para ilustrar o cálculo do lucro em um FIDC. Em cenários reais, há muitos outros fatores a serem considerados, como a distribuição dos pagamentos ao longo do tempo, a volatilidade dos ativos, as taxas de juros variáveis entre outros.

```python
import matplotlib.pyplot as plt
import numpy as np

# Valores de taxa de retorno
taxas_retorno = np.linspace(0, 0.1, 100)  # Varia de 0% a 20%

# Calcula o lucro para cada taxa de retorno
lucros = [calcular_lucro_fidc(valor_investido, taxa, taxa_administracao) for taxa in taxas_retorno]

# Cria o gráfico
plt.figure(figsize=(10, 6))
plt.plot(taxas_retorno, lucros)
plt.title('Variação do Lucro em um FIDC')
plt.xlabel('Taxa de Retorno')
plt.ylabel('Lucro')
plt.grid(True)
plt.show()
```

Este gráfico mostra como o lucro varia à medida que a taxa de retorno dos direitos creditórios varia. Lembre-se de que, em um cenário real, o gráfico seria mais complexo devido a outros fatores, como distribuição de pagamentos ao longo do tempo, flutuações do mercado financeiro entre outros.

## Vamos calcular a derivada da função que calcula o lucro de um FIDC em relação a cada um dos parâmetros

###Derivada em relação a valor_investido

```python
import sympy as sp

# Definindo os símbolos
valor_investido, taxa_retorno, taxa_administracao = sp.symbols('valor_investido taxa_retorno taxa_administracao')

# Definindo a função de lucro
lucro = (valor_investido * taxa_retorno) - (valor_investido * (taxa_administracao / 100))

# Calculando a derivada em relação a valor_investido
derivada_valor_investido = sp.diff(lucro, valor_investido)
print(f"derivada_valor_investido = {derivada_valor_investido}")
```

A derivada em relação a valor_investido nos dirá como o lucro varia quando o valor investido é alterado. Se a derivada for positiva, isso significa que um aumento no valor investido levará a um aumento no lucro, e vice-versa.

###Derivada em relação a taxa_retorno

```python
# Calculando a derivada em relação a taxa_retorno
derivada_taxa_retorno = sp.diff(lucro, taxa_retorno)
print(f"derivada_taxa_retorno = {derivada_taxa_retorno}")
```

A derivada em relação a taxa_retorno nos dirá como o lucro varia com mudanças na taxa de retorno. Se a derivada for positiva, isso significa que um aumento na taxa de retorno levará a um aumento no lucro.

###Derivada em relação a taxa_administracao

```python
# Calculando a derivada em relação a taxa_administracao
derivada_taxa_administracao = sp.diff(lucro, taxa_administracao)
print(f"derivada_taxa_administracao = {derivada_taxa_administracao}")
```

A derivada em relação a taxa_administracao nos dirá como o lucro varia com mudanças na taxa de administração. Se a derivada for negativa, isso significa que um aumento na taxa de administração levará a uma diminuição no lucro.

##Utilizando os valores do exemplo:


###*   valor_investido = 100000
###*   taxa_retorno = 0.08
###*   taxa_administracao = 1.5




```python
derivada_valor_investido = derivada_valor_investido.subs({valor_investido: 100000, taxa_retorno: 0.08, taxa_administracao: 1.5})
print(f"Derivada em relação a valor_investido: {derivada_valor_investido:.3f}")
```

A derivada em relação a valor_investido com os valores do exemplo é aproximadamente 0.065. Isso significa que, com um aumento de $1 no valor investido, o lucro aumentará em cerca de 0.065.

```python
derivada_taxa_retorno = derivada_taxa_retorno.subs({valor_investido: 100000, taxa_retorno: 0.08, taxa_administracao: 1.5})
print("Derivada em relação a taxa_retorno:", derivada_taxa_retorno)
```

A derivada em relação a taxa_retorno com os valores do exemplo é 10000. Isso significa que um aumento de 0.01 (ou 1%) na taxa de retorno resultará em um aumento de aproximadamente $10000 no lucro.

```python
derivada_taxa_administracao = derivada_taxa_administracao.subs({valor_investido: 100000, taxa_retorno: 0.08, taxa_administracao: 1.5})
print("Derivada em relação a taxa_administracao:", derivada_taxa_administracao)
```

A derivada em relação a taxa_administracao com os valores do exemplo é -1000. Isso significa que um aumento de 1% na taxa de administração resultará em uma diminuição de aproximadamente $1000 no lucro.
